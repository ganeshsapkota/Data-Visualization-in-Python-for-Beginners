```python
import matplotlib.pyplot as plt
import numpy as np

axis  = plt.axes(projection="3d")

x = np.arange(-2,2,0.02)
y  = np.arange(-2,2,0.02)

meshx, meshy = np.meshgrid(x,y)
meshz = np.cos(meshx) * np.cos(meshy)

axis.plot_surface(meshx,meshy,meshz, cmap="plasma")
axis.view_init(azim=8, elev=60)
plt.show()
```
