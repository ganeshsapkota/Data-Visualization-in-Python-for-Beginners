```python
import matplotlib.pyplot as plt

plt.style.use("fivethirtyeight")

hour = [1,2,3,4,5,6]

contestant1 = [1,2,3,4,5,6]
contestant2 = [1,2,3,6,5,7]
contestant3 = [1,1,4,6,5,8]

labels = ["Con 1", "Con 2", "Con 3"]

plt.stackplot(hour, contestant1, contestant2, contestant3, labels=labels)
plt.legend(loc="upper left")
plt.show()
```
